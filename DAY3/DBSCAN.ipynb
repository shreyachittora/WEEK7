```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering

```

```python
data = pd.read_csv("Mall_Customers.csv")

data.head()
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values

```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#####  DBSCAN


```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

plt.scatter(X[:, 0], X[:, 1], c=db_labels, cmap='rainbow', s=60)
plt.title("DBSCAN Clustering (Noise = -1)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1–100)")
plt.grid(True)

```

###### K-Means


```python
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans_labels = kmeans.fit_predict(X_scaled)

plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='rainbow', s=60)
plt.title("K-Means Clustering (k=5)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1–100)")
plt.grid(True)

```

##### Hierarchical Clustering


```python

agglo = AgglomerativeClustering(n_clusters=5, linkage='ward')
agglo_labels = agglo.fit_predict(X_scaled)

plt.scatter(X[:, 0], X[:, 1], c=agglo_labels, cmap='rainbow', s=40)
plt.title("Hierarchical Clustering (k=5)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1–100)")
plt.grid(True)
```
